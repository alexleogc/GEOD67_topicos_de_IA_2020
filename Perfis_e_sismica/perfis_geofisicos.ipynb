### Construindo um DataFrame

```python
#Utilizando dictionary comprehension: {key:value for (key,value) in dictonary.items()}
df = pd.DataFrame({i:well[i] for i in well.curves.keys()})
```

### Plotando perfis

```python
print(well.curves)
```

```python
df[(df.DEPT>=1000) & (df.DEPT<=2000)].describe()[['DEPT','GRGC','R40O','R60O','RTAO','DEN','NPRL','PDPE']]
```

```python
#Criando quatro quadros para plotagem dos perfis
fig,axes = plt.subplots(ncols=4,figsize=(10,12))

#Configurações gerais
for ax in axes:
    ax.set_ylim(1000,2000) #limitando eixo y
    ax.invert_yaxis() #invertendo eixo y
    ax.grid(which='both',ls='--')

#Primeiro plot
axes[0].plot(df.GRGC,df.DEPT,c='black',label='GR')
axes[0].set_xlim(0,250)
axes[0].set_title('Raios Gama')
axes[0].set_xlabel('uAPI')

#Segundo plot
axes[1].semilogx(df.R40O,df.DEPT,c='black',label='RT40')
axes[1].semilogx(df.R60O,df.DEPT,c='blue',ls='--',label='RT60')
axes[1].semilogx(df.RTAO,df.DEPT,c='red',ls='-.',label='RT')
axes[1].set_xlim(0.2,20)

#Terceiro plot
axes[2].plot(df.DEN,df.DEPT,c='black',label='Dens')
axes[2].set_xlim(2,3)
ax2 = axes[2].twiny() #Criando um eixo gêmeo
ax2.plot(df.NPRL,df.DEPT,c='red',ls='--',label='Neut')
ax2.set_xlim(-15,60)

#Quarto plot
axes[3].plot(df.PDPE,df.DEPT,label='PE',color='black')
axes[3].set_xlim(1,5)

for ax in axes:
    ax.legend()
fig.tight_layout()
```
