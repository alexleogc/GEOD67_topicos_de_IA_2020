```python
#Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error 
from Gradiente import *
```

### Criando um dataset sintético

```python
x1 = np.random.normal(scale=1,size=50)

y = x1+np.random.normal(scale=0.4,size=50)
```

```python
fig, axes = plt.subplots()

axes.scatter(x1,y)

fig.tight_layout()
```

```python
np.corrcoef(x1,y)
```

### Regressão utilizando o método do gradiente (ver slide)

```python
X = add_vetor_unit(x1)
X
```

#### Criando modelo inicial nulo

```python
coef_ini = np.asarray([0,0])
coef_ini
```

```python
x_n = np.linspace(-3,3,100)
```

```python
fig, axes = plt.subplots()

axes.scatter(x1,y)

axes.plot(x_n,predict(coef_ini,add_vetor_unit(x_n)))

fig.tight_layout()
```

#### Obtendo os coeficientes da regressão

```python
coef_grad, cost_hist, coef_hist = gradient_descent(X,y,coef_ini,alpha=0.02)
```

#### Plotando função custo

```python
plt.plot(cost_hist)
```

```python
coef_grad
```

```python
fig, axes = plt.subplots()

axes.scatter(x1,y)

axes.plot(x_n,predict(coef_grad,add_vetor_unit(x_n)))

fig.tight_layout()
```

```python
mean_squared_error(y,predict(coef_grad,X))
```
